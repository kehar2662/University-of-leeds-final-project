## Data import and organized

```python
# import the packages needed in this file
import pandas as pd
import numpy as np
```

```python
# read all file into dataframe and show all columns
bf_summary=pd.read_csv('brief_summaries.txt', sep='|', index_col=False)
bc=pd.read_csv('browse_conditions.txt', sep='|', index_col=False)
bi=pd.read_csv('browse_interventions.txt', sep='|', index_col=False)
countries=pd.read_csv('countries.txt', sep='|', index_col=False)
detail=pd.read_csv('detailed_descriptions.txt', sep='|', index_col=False)
eligibilities=pd.read_csv('eligibilities.txt', sep='|', index_col=False)
facilities=pd.read_csv('facilities.txt', sep='|', index_col=False)
interventions=pd.read_csv('interventions.txt', sep='|', index_col=False)
facility_contacts=pd.read_csv('facility_contacts.txt', sep='|', index_col=False)
keywords=pd.read_csv('keywords.txt', sep='|', index_col=False)
sponsors=pd.read_csv('sponsors.txt', sep='|', index_col=False)
detail=pd.read_csv('detailed_descriptions.txt', sep='|', index_col=False)
studies=pd.read_csv('studies.txt', sep='|', index_col=False, low_memory=False)
pd.set_option('display.max_columns', None)
```

```python
# remove all unnesscary columns and show the dataframe after filter
columns_to_keep = ['nct_id', 'description']
bf_summary_filtered = bf_summary[columns_to_keep]
bf_summary_filtered
```

```python
# remove all unnesscary columns and show the dataframe after filter
# the cell collapse into one if they have the same nct_id value
columns_to_keep = ['nct_id', 'downcase_mesh_term']
bc_filtered = bc[columns_to_keep]
bc_filtered = bc_filtered.groupby('nct_id').agg(lambda x: ', '.join(x.dropna().astype(str).unique())).reset_index()
bc_filtered
```

```python
# remove all unnesscary columns and show the dataframe after filter, the columns name will be renamed to avoid the same columns name
# the cell collapse into one if they have the same nct_id value
columns_to_keep = ['nct_id', 'downcase_mesh_term']
bi_filtered = bi[columns_to_keep]
bi_filtered_copy = bi_filtered.copy()
bi_filtered_copy.rename(columns={'downcase_mesh_term': 'interventions_mesh_term'}, inplace=True)
bi_filtered_copy = bi_filtered_copy.groupby('nct_id').agg(lambda x: ', '.join(x.dropna().astype(str).unique())).reset_index()
bi_filtered_copy
```

```python
# remove all unnesscary columns and show the dataframe after filter, the columns name will be renamed to avoid the same columns name
# the cell collapse into one if they have the same nct_id value
columns_to_keep = ['nct_id', 'name']
countries_filtered = countries[columns_to_keep]
countries_filtered_copy = countries_filtered.copy()
countries_filtered_copy.rename(columns={'name': 'countries_name'}, inplace=True)
countries_filtered_copy = countries_filtered_copy.groupby('nct_id').agg(lambda x: ', '.join(x.dropna().astype(str).unique())).reset_index()
countries_filtered_copy
```

```python
# remove all unnesscary columns and show the dataframe after filter, the columns name will be renamed to avoid the same columns name
# the cell collapse into one if they have the same nct_id value
columns_to_keep = ['nct_id', 'description']
detail_filtered = detail[columns_to_keep]
detail_filtered_copy = detail_filtered.copy()
detail_filtered_copy.rename(columns={'description': 'detailed_description'}, inplace=True)
detail_filtered_copy = detail_filtered_copy.groupby('nct_id').agg(lambda x: ', '.join(x.dropna().astype(str).unique())).reset_index()
detail_filtered_copy
```

```python
# remove all unnesscary columns and show the dataframe after filter
columns_to_keep = ['nct_id', 'gender','minimum_age','maximum_age','criteria']
eligibilities_filtered = eligibilities[columns_to_keep]
eligibilities_filtered
```

```python
# remove all unnesscary columns and show the dataframe after filter, the columns name will be renamed to avoid the same columns name
# the cell collapse into one if they have the same nct_id value
columns_to_keep = ['nct_id', 'name','city','state','country']
facilities_filtered = facilities[columns_to_keep]
facilities_filtered_copy = facilities_filtered.copy()
facilities_filtered_copy.rename(columns={'name': 'facility_name'}, inplace=True)
facilities_filtered_copy = facilities_filtered_copy.groupby('nct_id').agg(lambda x: ', '.join(x.dropna().astype(str).unique())).reset_index()
facilities_filtered_copy
```

```python
# remove all unnesscary columns and show the dataframe after filter, the columns name will be renamed to avoid the same columns name
# the cell collapse into one if they have the same nct_id value
columns_to_keep = ['nct_id', 'name','email','phone','phone_extension']
facility_contacts_filtered = facility_contacts[columns_to_keep]
facility_contacts_filtered_copy = facility_contacts_filtered.copy()
facility_contacts_filtered_copy.rename(columns={'name': 'contact_name'}, inplace=True)
facility_contacts_filtered_copy
```

```python
# remove all unnesscary columns and show the dataframe after filter, the columns name will be renamed to avoid the same columns name
# the cell collapse into one if they have the same nct_id value
columns_to_keep = ['nct_id', 'downcase_name']
keywords_filtered = keywords[columns_to_keep]
keywords_filtered_copy = keywords_filtered.copy()
keywords_filtered_copy.rename(columns={'downcase_name': 'keywords'}, inplace=True)
keywords_filtered_copy = keywords_filtered_copy.groupby('nct_id').agg(lambda x: ', '.join(x.dropna().astype(str).unique())).reset_index()
keywords_filtered_copy
```

```python
# remove all unnesscary columns and show the dataframe after filter, the columns name will be renamed to avoid the same columns name
# the cell collapse into one if they have the same nct_id value
columns_to_keep = ['nct_id','agency_class', 'name']
sponsors_filtered = sponsors[columns_to_keep]
sponsors_filtered_copy = sponsors_filtered.copy()
sponsors_filtered_copy.rename(columns={'name': 'agency_name'}, inplace=True)
sponsors_filtered_copy = sponsors_filtered_copy.groupby('nct_id').agg(lambda x: ', '.join(x.dropna().astype(str).unique())).reset_index()
sponsors_filtered_copy
```

```python
# remove all unnesscary columns and show the dataframe after filter
columns_to_keep = ['nct_id','study_type', 'official_title','overall_status','phase','source',
                   'number_of_arms','is_fda_regulated_drug','is_fda_regulated_device']
studies_filtered = studies[columns_to_keep]
studies_filtered
```

```python
# remove all unnesscary columns and show the dataframe after filter, the columns name will be renamed to avoid the same columns name
# the cell collapse into one if they have the same nct_id value
columns_to_keep = ['nct_id','intervention_type', 'name']
interventions_filtered = interventions[columns_to_keep]
interventions_filtered_copy = interventions_filtered.copy()
interventions_filtered_copy.rename(columns={'name': 'drug_name'}, inplace=True)
interventions_filtered_copy = interventions_filtered_copy.groupby('nct_id').agg(lambda x: ', '.join(x.dropna().astype(str).unique())).reset_index()
interventions_filtered_copy
```

```python
#combine all dataframe into one and reveal the combined dataframe
combined_df = pd.concat([bf_summary_filtered, bc_filtered, bi_filtered_copy, countries_filtered_copy,detail_filtered_copy,
                         eligibilities_filtered, facilities_filtered_copy, facility_contacts_filtered_copy,keywords_filtered_copy,
                         sponsors_filtered_copy,studies_filtered,interventions_filtered_copy])
combined_df
```

```python
# collapse the row into one if they have the same NCT_ID value
combined_df_unique = combined_df.groupby('nct_id').sum().reset_index()
combined_df_unique
```

## Data cleaning

```python
# Check empty value within each column
per_column = (combined_df_unique == 0).sum()
per_column
```

```python
# clean up the empty value.
# The row without description,countries_name, criteria,facility_name, study_type, official_title will be deleted due to those info cannot be replaced
# The row with gender missing will be replaced by all 
# The row with minimum_age,maximum_age missing will be replaced by 0 and 120 
# The row with other columns missing will be replaced by n.a or unknown
combined_df_unique_filtered = combined_df_unique[combined_df_unique['description'] != 0]
combined_df_unique_filtered = combined_df_unique_filtered[combined_df_unique['countries_name'] != 0]
combined_df_unique_filtered = combined_df_unique_filtered[combined_df_unique['criteria'] != 0]
combined_df_unique_filtered = combined_df_unique_filtered[combined_df_unique['facility_name'] != 0]
combined_df_unique_filtered = combined_df_unique_filtered[combined_df_unique['study_type'] != 0]
combined_df_unique_filtered = combined_df_unique_filtered[combined_df_unique['official_title'] != 0]
combined_df_unique_filtered['gender'] = combined_df_unique_filtered['gender'].replace(0, 'All')
combined_df_unique_filtered['downcase_mesh_term'] = combined_df_unique_filtered['downcase_mesh_term'].replace(0, 'n.a')
combined_df_unique_filtered['interventions_mesh_term'] = combined_df_unique_filtered['interventions_mesh_term'].replace(0, 'n.a')
combined_df_unique_filtered['detailed_description'] = combined_df_unique_filtered['detailed_description'].replace(0, 'n.a')
combined_df_unique_filtered['minimum_age'] = combined_df_unique_filtered['minimum_age'].replace(0, '0')
combined_df_unique_filtered['maximum_age'] = combined_df_unique_filtered['maximum_age'].replace(0, '120 Years')
combined_df_unique_filtered['contact_name'] = combined_df_unique_filtered['contact_name'].replace(0, 'unknown')
combined_df_unique_filtered['email'] = combined_df_unique_filtered['email'].replace(0, 'unknown')
combined_df_unique_filtered['phone'] = combined_df_unique_filtered['phone'].replace(0, 'unknown')
combined_df_unique_filtered['phone_extension'] = combined_df_unique_filtered['phone_extension'].replace(0, 'unknown')
combined_df_unique_filtered['keywords'] = combined_df_unique_filtered['keywords'].replace(0, 'n.a')
combined_df_unique_filtered['phase'] = combined_df_unique_filtered['phase'].replace(0, 'unknown')
combined_df_unique_filtered['is_fda_regulated_device'] = combined_df_unique_filtered['is_fda_regulated_device'].replace(0, 'unknown')
combined_df_unique_filtered['is_fda_regulated_drug'] = combined_df_unique_filtered['is_fda_regulated_drug'].replace(0, 'unknown')
combined_df_unique_filtered['number_of_arms'] = combined_df_unique_filtered['number_of_arms'].replace(0, 'unknown')
combined_df_unique_filtered['intervention_type'] = combined_df_unique_filtered['intervention_type'].replace(0, 'unknown')
combined_df_unique_filtered['drug_name'] = combined_df_unique_filtered['drug_name'].replace(0, 'unknown')
```

```python
# reset the index number and show the dataframe after cleanup
combined_df_unique_filtered=combined_df_unique_filtered.reset_index(drop=True)
combined_df_unique_filtered
```

```python
# make sure empty has been cleaned
per_column_after = (combined_df_unique_filtered == 0).sum()
per_column_after
```

```python
# check how many different possible result of clinical trial
unique_values = combined_df_unique_filtered['overall_status'].unique()
unique_values
```

```python
# Terminated, 'Withdrawn,Unknown status, Suspended ,No longer available will be markerd as fail (0)
# 'Completed', 'Approved for marketing','Available' will be markerd as succeed (0)
values_to_replace = ['Terminated', 'Withdrawn','Unknown status','Suspended','No longer available']
values_to_replace2 =['Completed', 'Approved for marketing','Available']
combined_df_unique_filtered['overall_status'] = np.where(combined_df_unique_filtered['overall_status'].isin(values_to_replace),  0,
                                                         combined_df_unique_filtered['overall_status'])
combined_df_unique_filtered['overall_status'] = np.where(combined_df_unique_filtered['overall_status'].isin(values_to_replace2),  1,
                                                         combined_df_unique_filtered['overall_status'])
combined_df_unique_filtered
```

```python
# check how many different possible result of clinical trial after replacement
unique_values = combined_df_unique_filtered['overall_status'].unique()
unique_values
```

```python
unique_values = combined_df_unique_filtered['countries_name'].unique()
unique_values
```

```python
# check how many trial are "Temporarily not available"
count_ongoing_trial = combined_df_unique_filtered['overall_status'].isin([ "Temporarily not available"]).sum()
count_ongoing_trial
```

```python
# remove all Temporarily not available since they cannot be classified as success or fail or ongoing. This only reduce 15 rows
combined_df_unique_filtered = combined_df_unique_filtered[combined_df_unique_filtered['overall_status'] != "Temporarily not available"]
unique_values = combined_df_unique_filtered['overall_status'].unique()
unique_values
```

```python
null_counts = combined_df_unique_filtered.isnull().sum()
null_counts
```

```python
#split the completed trials and ongoing trials into two files for fruther usage
condition_value=[0,1]
completed_trial= combined_df_unique_filtered[combined_df_unique_filtered['overall_status'].isin(condition_value)]
ongoing_trial= combined_df_unique_filtered[~combined_df_unique_filtered['overall_status'].isin(condition_value)]
completed_trial.to_csv('completed_trial.csv', index=False)
ongoing_trial.to_csv('ongoing_trial.csv', index=False)
```
