```python
# import the packages needed in this file
import pandas as pd
import numpy as np
```

```python
# Import the ongoing clincial trial data as dataframe 
ongoing_trial=pd.read_csv('ongoing_trial.csv', index_col=False)
ongoing_trial
```

```python
# Make a copy of dataframe so that the following operation will not affect the orignal data
ongoing_trial_copy=ongoing_trial.copy()
```

```python
#Since the age has been expressed in Years, Months, Weeks, Days, Hours, I write a function to transform the age unit into years.
def convert_to_years(age_str):
    """This function will transform the time's unit from Months, Weeks, Days, Hours into Years """
    num = float(age_str.split()[0])
    if 'Year' in age_str:
        return num
    elif 'Month' in age_str:
        return num / 12  
    elif 'Week' in age_str:
        return num / 52  
    elif 'Day' in age_str:
        return num / 365  
    elif 'Hour' in age_str:
        return num / (365 * 24) 
    else:
        return 0
```

```python
ongoing_trial_copy['minimum_age'] = ongoing_trial_copy['minimum_age'].apply(convert_to_years)
ongoing_trial_copy['maximum_age'] = ongoing_trial_copy['maximum_age'].apply(convert_to_years)
ongoing_trial_copy
```

```python
null_counts = ongoing_trial_copy.isnull().sum()
null_counts
```

```python
ongoing_trial_copy = ongoing_trial_copy.dropna(subset=['facility_name'])
ongoing_trial_copy = ongoing_trial_copy.dropna(subset=['drug_name'])
null_counts = ongoing_trial_copy.isnull().sum()
null_counts
```
